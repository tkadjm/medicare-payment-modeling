# The Notebook Will Process the Raw Medicare Files
We need two Medicare Part B payment files to do the simulation of the proposed Medicare Physician Fee Schedule. The first two variables are the locations of the files on the Web. Download and unzip both files. When the notebook as been run, there will be two files named with the other two variables. The next script will need those file names, so if you change them, you will need to change them in the other script(s).

```python
medicareFileUrl = "http://www.cms.gov/apps/ama/license.asp?file=https://downloads.cms.gov/files/Medicare-Physician-and-Other-Supplier-PUF.zip"
medicareAggFileUrl = "https://downloads.cms.gov/files/Medicare-Physician-and-Other-Supplier-NPI-Aggregate_CY2016.zip"

medicare_saved = "medicare_saved.pkl"
medicare_agg_saved = "medicare_agg_saved.pkl"
```

After you have downloaded and unzip the files. You will need to tell this notebook where those files are located. Replace the current locations with the locations of the files on your computer. If you put them in the same folder, you will not need to put the full path to the files.

```python
medicareFileName = "/Users/swaldren/OG/ProviderData/PartB/Medicare_Provider_Util_Payment_PUF_CY2016.txt"
medicareAggFileName = "/Users/swaldren/OG/ProviderData/PartB/Medicare_Physician_and_Other_Supplier_NPI_Aggregate_CY2016.txt"
```

```python
import pandas as pd
```

```python
df_medicare = pd.read_csv(medicareFileName, sep='\t', header=0, skiprows=[1])
df_medicare.set_index('NPI')
df_medicare.head()
```

Since the Medicare files are large, the following filters out only the columns we care about.

```python
df_medicare_filter = df_medicare[['NPI', 'PLACE_OF_SERVICE', 'HCPCS_CODE', 'LINE_SRVC_CNT', 'PROVIDER_TYPE']]
df_medicare_filter.to_pickle(medicare_saved)
```

```python
df_medicare_agg = pd.read_csv(medicareAggFileName, sep='\t')
df_medicare_agg.set_index('NPI')
df_medicare_agg.head()
```

Since the Medicare files are large, the following filters out only the columns we care about.

```python
df_medicare_agg_filter = df_medicare_agg[['NPI', 'NPPES_ENTITY_CODE', 'BENEFICIARY_AVERAGE_RISK_SCORE', 'NPPES_PROVIDER_STATE',
                                  'TOTAL_UNIQUE_BENES', 'PROVIDER_TYPE']]
df_medicare_agg_filter.to_pickle(medicare_agg_saved)
```
